```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot
```

```python
def SentToken(sentences):
    dataset = sentences
    Sentences = [sentences.split() for sentences in dataset]
    print(Sentences)
    return Sentences
```

```python
engSentences = SentToken(english_sentences)
```

```python
TamSentences = SentToken(tamil_sentences)
```

```python
def OwnwordModel(langsentences,modelname):
    from gensim.models import Word2Vec
    from sklearn.decomposition import PCA
    import matplotlib.pyplot as plt
    model = Word2Vec(langsentences, min_count=1)
    print(model)
    
    model.save(modelname)

    X=model.wv.vectors
    pca = PCA(n_components=2)
    result = pca.fit_transform(X)

    plt.scatter(result[:, 0], result[:, 1])
    words = list(model.wv.index_to_key)
    for i, word in enumerate(words):
        plt.annotate(word, xy=(result[i, 0], result[i, 1]))
    plt.show()
    
```

```python
OwnwordModel(engSentences,"engmodel.bin")
```
